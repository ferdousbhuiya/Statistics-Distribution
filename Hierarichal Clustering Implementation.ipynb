```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
# Standarization of Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
# Now we have to reduce the feature by using the PCA

from sklearn.decomposition import PCA
# Here we are converting 4 features into 2 features
pca = PCA(n_components=2)
pca.fit_transform(X_scaled)
```

### The above data shows that it converts to 2 features only. 

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=iris.target) 
```

```python
# AggLomerative Clustering
# To Construct a dendogram
import scipy.cluster.hierarchy as sc

# plot the dendogram

plt.figure(figsize=(20,7))
plt.title("Dendogram")

# create the Dendogram
sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
plt.title("Dendogram")
plt.xlabel("Sample")
plt.ylabel("Ucladian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity="euclidean", linkage="ward")
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_) 
```

### From this above figure we can say that it easily converts to 2 groups from the 3 groups of features.
