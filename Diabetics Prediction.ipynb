```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
import pandas as pd 
df_diabets = pd.DataFrame(dataset.data, columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
```

```python
df_diabets
```

```python
# Independant and Dependant features
X = df_diabets
y= dataset['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=10)
```

```python
X_train.head()
```

```python
# Check the correlation
X_train.corr()
```

```python
# heatmap of correlation
import matplotlib.pyplot as plt
import seaborn as sns

# If using Jupyter notebook, keep this:
# %matplotlib inline

# Create a figure with specific dimensions
plt.figure(figsize=(15, 10))

# Generate a heatmap with annotations
sns.heatmap(X_train.corr(), annot=True, cmap='coolwarm', fmt=".2f")

# Show the plot
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

## Hyperparameter Tuning


```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','log2','sqrt']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(regressor, param_grid=param, cv=5, scoring='neg_mean_squared_error', )
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
#prediction
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

## Let us see the Decision Tree as per our param

```python
selectedmodel1 = DecisionTreeRegressor(criterion='friedman_mse', max_depth=4,max_features='sqrt', splitter='best' )
```

```python
selectedmodel1.fit(X_train, y_train)
```

```python
# Visualize the decission Tree
from sklearn import tree
import matplotlib.pyplot as plt

# Adjust the figure size
plt.figure(figsize=(15,10))

# Plot the decision tree (assuming 'selectedmodel1' is a fitted decision tree model)
tree.plot_tree(selectedmodel1, filled=True)

# Show the plot
plt.show()
```

```python
y_pred1 = selectedmodel1.predict(X_test)
```

```python
print(r2_score(y_test,y_pred1))
print(mean_absolute_error(y_test,y_pred1))
print(mean_squared_error(y_test,y_pred1))
```

# Let us see the performance as per our selected params

```python
selectedmodel2 = DecisionTreeRegressor(criterion='squared_error', max_depth=3,max_features='sqrt', splitter='best' )
selectedmodel2.fit(X_train, y_train)
```

```python
# Adjust the figure size
plt.figure(figsize=(15,10))

# Plot the decision tree (assuming 'selectedmodel1' is a fitted decision tree model)
tree.plot_tree(selectedmodel2, filled=True)

# Show the plot
plt.show()
```

```python
y_pred2 = selectedmodel2.predict(X_test)
print(r2_score(y_test,y_pred2))
print(mean_absolute_error(y_test,y_pred2))
print(mean_squared_error(y_test,y_pred2))
```
